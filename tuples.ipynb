
# Tuples in python



```python
tup = (1,2,3,4)
print(tup)
```

```python
tup1 = (1,2,3,4)
tup2 = (3,4,5,6)
tup3 = tup1 + tup2
tup4 = [*(tup3)]
print(tup4)
```

#index(elememt, start, end)

```python
print(tup4)

x = tup4.index(3)
y = tup4.index(3,3,6)
print(x)
print(y)
```

##count()
**syntax** : count(element)

```python
z = tup4.count(3)
print(z)
f = sum(tup4)
print(f)
```

#max()

```python
print(tup4)
t  =  max(tup4)
print(t)
```

#min()

```python
print(tup4)
t = min(tup4)
print(t)
```

##Sorted( Tuple_name )

```python
tup5 = (3,1,5,2,7,8)
print(sorted(tup5))
```
