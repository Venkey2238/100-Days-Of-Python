##Creation of the array 
###**syntax** : array(data_type,value_list)

```python
import array as arr
a = arr.array('i',[1, 2, 3, 4])
print(type(a))
print(type(a[1]))
print(a)
```

```python
import array as arr
a = arr.array('d',[1.2,2.3,3.6])
print(a)
print(type(a))
print(type(a[1]))
```

#accessing the elements of an array

```python
for i in range(len(a)):
  print(a[i])
```

#adding the elements to the array

```python
print(a)
a.insert(2,4.5)
a.append(5.6)
print(a)
```
