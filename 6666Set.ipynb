#First method of declaring of the set

```python
t = set((1,2,3,4))
print(t)
print(type(t))
```

#Second method for declaring the string 

```python
e = {1,2,3,4}
print(e)
print(type(e))
```

#adding the element to the set 

```python
print(e)
e.add(8)
e.add(10)
print(e)
```

# accessing the elements of the set

```python
print(e)
for i in e:
  print(i,"",end="")
```

#Removing an element from the set

```python
print(e)
e.remove(3)
print(e)
```

#removing all elements in the set

```python
r = set()
for i in e:
  r.add(i)
print(r)
r.clear()
print(r)
```

#pop() in set dont take any attributes

```python
s = {1,2,3}
s.pop()
print(s)
```

#copy() copy the set 

```python
print("s : ",s)
rt = s.copy()
print(rt)
print(type(rt))
```

# update()  adds the elemts of one set to another set

```python
print(rt)
er = {1,6}
print(s)
rt.update(er)
print(rt)
```

#union() operation in the sets
union also uses the "|" this operator to do the union operation 
"|" this operatoe is also called as the pipe operator 

```python
a = {1,2,3,4}
b = {4,5,6,7}
c = a.union(b)
d = a | b
print(c)
print(d)
```

#difference() 
this method will not update the a set so we have another methods for that

```python
a = {1,2,3,4}
b = {3,4,5,6}
t = a.difference(b)
print(a)
print(t)
```

#difference_update()

```python
a = {3,4,5,6}
b = {5,2,4,6}
a.difference_update(b)
print(a)
```

#intersection()

```python
a = {1,2,3,4}
b = {3,4,5,6}
t = a.intersection(b)
print(a)
print(t)
```

#intersection_update

```python
a = {1,2,3,4}
b = {4,5,6,7}
a.intersection_update(b)
print(a)
```

#isdisjoint() will return true if there is noo intersection element 

```python
a = {1,2,3}
b = {4,5,6}
c = {1,2,3}
d = {3,4,5}
t = a.isdisjoint(b)
r = c.isdisjoint(d)
print(t)
print(r)
```

#issubset() returns true if the give set is subset of other set specified

```python
a  = {1,2,3,4,5}
b = {1,2}
s = b.issubset(a)
r = a.issuperset(b)
print(s)
print(r)
```

#symmetrical_difference() returns the uncommon elements in the set

```python
a = {1,2,3}
b = {3,4,5}
t = a.symmetric_difference(b)
print(t)
```

#symmetric_difference_update()

```python
a = {1,2,3}
b = {3,4,5}
a.symmetric_difference_update(b)
print(a)
print(b)
```
