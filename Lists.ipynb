##Slicing a list
* [:index] === To print elements from beginning to a range use
* [:-index] === To print elements from end-use
* [index:] === To print elements from a specific index till the end use 
* [::-1 ] === prints the reverse order
* [1::2] ===  prints the letter or number in the list at the index 1 and and jumps to step to 2 from index 1

means assume list[1,2,3,4] [1::2]  [intial:End:Jump] ====> returns [2,4]


```python
x = [1,2,3,4,5]
print(x[:2])
print(x[:-1])
print(x[2:])
print(x[::-1])
print(x[1::3])
print(x[1:2])
print(x[:])  ### prints all the elements
```

#Q) for first n natural numbers,find all distinct pairs of numbers whose product  is odd 

```python
n = int(input("enter the number of natual numbers "))
x = []
for i in range(1,n):
  for j in range(1,n):
    if (i*j % 2) == 1:
      x.append((i,j))
print(x)
```

```python
n = int(input("enter the number of natual numbers "))
x = []
for i in range(1,n):
  for j in range(i+1,n):
    if (i*j % 2) == 1:
      x.append((i,j))
print(x)
```

# filter() function in python

Syntax : filter(function, iterable)

Filter checks the given lambda function accpets the element in a range(1,n)

```python
seq = [0, 1, 2, 3, 5, 8, 13]
  
# result contains odd numbers of the list
result = filter(lambda x: x % 2 != 0, seq)
print(list(result))
```

```python
list1 = filter(lambda x:x%2 == 1, range(1,20))
print(list(list1))
```

```python
from functools import reduce

scores = [75, 65, 80, 95, 50]

total = reduce(lambda a, b: a + b, scores)

print(total)
```
