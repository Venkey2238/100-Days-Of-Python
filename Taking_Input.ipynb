### Given a list [1,2,3,4,5] you need to multilply each element by 5 and print a new list using the list comprehension


```python
##general method without list  list comprehension
list1 = [1,2,3,4,5]
list2 = []
x = len(list1)
y = list1.count(1)
print(y)
print(x)
for i in range(0,len(list1)):
  list2.append(list1[i]*5)
print(list2)
```

##In the above code i ahve specified two methods of the list they are length and count
Length -> return the total count that is for [1,2,3,4,5] the length is 5
Count -> this method is used to count the number of occurences in the list like in list [1,2,3,4,5] list.count(1) provides the number of times the 1 occured that is 1 time

```python
##using the list comprehension
list1  = [1,2,3,4,5]
list2 = [x*5 for x in list1 ]
print(list2)
```

in the above code see the list comprehension step read it form right to left for x in list 1 we are doing x*5 

## Give a list [1,2,3,4,5,6] print the even list and odd list using list comprehension

```python
list1 = [1,2,3,4,5,6]
even = [x for x in list1 if x%2==0]
odd = [x for x in list1 if x%2!=0]
print(even)
print(odd)
```

```python
r = int(input("enter lower range"))
t = int(input("enter upper range"))
print("-----------Before formating----------")
for i in range(r,t):
  print(i**1,i**2,i**3,i**4)
print("-----------After formating-----------")
for i in range(r,t):
  print("{:6d} {:6d} {:6d} {:6d}".format(i**1,i**2,i**3,i**4))
```

For the above code we have the :6d which means the spacing of the word in a senctences or for above we designed a space for numbers to provide a format for it

```python
txt = "hello my name is {name} i am {age} old"
hello = {
    'name' : 'venkey',
    'age' : 'twenty'
}
print(txt.format(**hello))
```
