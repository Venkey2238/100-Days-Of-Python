# **Nested Dictionary**

#Adding element to the dictionary

```python
Dict1 = {}
Dict1[0] = "hello"
Dict1[1] = "geeks"
Dict1[2] = 20
Dict1[3] = 30
print(Dict1)
```

#Accessing the elements of the Dictionary

```python
print(Dict1)
d = Dict1.get(2)
t = Dict1.get(0)
print(d)
print(t)
```

# Accesing a nested a list 


##**syntax** : Dictonaryname['key of nested dictionar']['key inside the nested dictionary']
###Example is given below

```python
Dict = {
   1 : {1:"hello",2:"geeks"},
   2 : {1:"go",2:"recive"}
}
print(Dict[1])
print("----------------------")
print(Dict)
print("Dict[1][2] : ",Dict[1][2])
print("-----------------------")
print(Dict)
print("Dict[2][1] : ",Dict[2][1])
```

```python
Dict3 = Dict.copy()
print(Dict3)
Dict3.clear()
print(Dict3)
```

```python
print(Dict.items())
print(Dict.keys())
```
