```python
N = int(input())
L = []
for i in range(0,N):
 x = [x for x in input().split()]
 if x[0]=="insert":
  L.insert(int(x[1]),int(x[2]))
 if x[0]=="append":
   L.append(int(x[1]))
 if x[0]=="remove":
   L.remove(x[1])
 if x[0]=="print":
   print(L)
 if x[0] == "pop":
   L.pop()
 if x[0] == "sort":
   L.sort()
 if x[0]=="reverse":
   L.reverse()
```
